```python
url = 'https://www.goszakup.gov.kz/ru/registry/rqc'
request = requests.get(url, verify=False)
```

```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(request.text, 'lxml')
```

```python
s.find('div', class_='table-responsive').findAll('tr')[1].text
```

```python
s.findAll('tr')
```

```python
contrs = s.findAll('div', class_='table-responsive')

dat = []

for contr in contrs: 
    name = contr.findAll('td')
    
    
    dat.append(name)
```

```python
dat
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dat)
```
